```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 8
x = np.linspace(0, 8, 1000) # Observe the signals of [0, 8] seconds. Sampling rate: 100 Hz.

def create_signal(k):
    theta = 2*np.pi*k*x
    s = np.cos(theta) + 1j*np.sin(theta)
    return s
```

```python
s1 = create_signal(1)
s2 = create_signal(2)
s3 = create_signal(3)
```

```python
def visualization(s, k):
    reals = [z.real for z in s]
    imags = [z.imag for z in s]
    
    with plt.style.context('classic'):
        plt.figure(figsize=(8, 1), dpi=100)
        plt.plot(x, reals, label='Real',color='red',linewidth=1)
        plt.plot(x, imags, label='Imaginary',color='green',linewidth=1)
        
        plt.xlabel('Time(s)')
        plt.ylabel('Amplitude')
        plt.xlim(0, 8)
        plt.ylim(-1.5, 1.5)
        plt.legend(bbox_to_anchor=(1.3, 0.5))
        plt.title('K='+str(k))
        plt.show()
```

```python
visualization(s1, 1)
visualization(s2, 2)
visualization(s3, 3)
```
